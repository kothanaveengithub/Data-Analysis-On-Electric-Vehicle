### Dataset Description

The dataset comprises information on electric vehicles (EVs) available in the market, including various attributes such as:

- **Make and Model:** The manufacturer and specific model of the vehicle.
- **Model Year:** The year the vehicle was produced.
- **Electric Vehicle Type:** Classification of the vehicle (e.g., Battery Electric Vehicle (BEV), Plug-in Hybrid Electric Vehicle (PHEV)).
- **Electric Range:** The maximum distance the vehicle can travel on electric power alone.
- **Base MSRP:** The manufacturer's suggested retail price.
- **Location Data:** Includes city, state, and postal code information, which can be used for geographical analysis.

### Insights from the Dataset

1. **Market Trends:** Analyzing the dataset can reveal trends in the adoption of different EV makes and models over the years, indicating which brands are gaining popularity.

2. **Geographic Distribution:** By mapping the data, one can identify regions with higher EV adoption, providing insights into market penetration and potential areas for growth.

3. **Vehicle Types Analysis:** Insights into the types of electric vehicles prevalent in the dataset can help understand consumer preferences, whether leaning more towards fully electric vehicles or hybrids.

4. **Price Range Insights:** The analysis of base MSRP can provide valuable information about the pricing strategies of different manufacturers and how it correlates with the vehicle's electric range and type.

5. **Environmental Impact:** By examining the electric range of various vehicles, one can infer the potential environmental benefits of EV adoption in different areas, which is crucial for policymakers and environmentalists.

This dataset serves as a rich resource for understanding the electric vehicle market, aiding stakeholders in making informed decisions regarding production, marketing, and infrastructure development.

## Handling missing values

```python
# Fill missing values for 'Model' with 'Unknown'
data['Model'].fillna('Unknown', inplace=True)

# For 'Legislative District', fill with the mode (most common value)
data['Legislative District'].fillna(data['Legislative District'].mode()[0], inplace=True)

# For 'Vehicle Location', fill with 'Unknown'
data['Vehicle Location'].fillna('Unknown', inplace=True)

# For 'Electric Utility', fill with 'Unknown'
data['Electric Utility'].fillna('Unknown', inplace=True)

# Check for any remaining missing values
print(data.isnull().sum())
```

## Outliers detection

```python
# List of numerical columns
numerical_columns = ['Model Year', 'Electric Range', 'Base MSRP']

# Set the style for the plots
sns.set(style="whitegrid")

# Create boxplots for each numerical column to visualize outliers
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

## Removing outliers

```python
# List of numerical columns to remove outliers from
numerical_columns = ['Model Year', 'Electric Range', 'Base MSRP']

# Function to remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for the specified numerical columns
for col in numerical_columns:
    data = remove_outliers(data, col)

# Display the new summary statistics to verify
print(data.describe())
```

## Task2 : Create a Choropleth using plotly.express to display the number of EV vehicles based on location.

```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv(r"C:\Users\navee\Downloads\dataset.csv")

# Step 1: Aggregate the data
# Count the number of EVs by state
ev_counts_by_state = data['State'].value_counts().reset_index()
ev_counts_by_state.columns = ['State', 'EV_Count']

# Step 2: Create the Choropleth map
choropleth_fig = px.choropleth(ev_counts_by_state, 
                                 locations='State', 
                                 locationmode='USA-states',  # for US states
                                 color='EV_Count',
                                 scope='usa',  # Limit map to the USA
                                 title='Number of Electric Vehicles by State',
                                 color_continuous_scale=px.colors.sequential.Plasma,
                                 labels={'EV_Count': 'Number of EVs'})
# Show the figure
choropleth_fig.show()
```
